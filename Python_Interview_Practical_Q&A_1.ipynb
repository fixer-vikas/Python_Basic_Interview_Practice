```python
# Insert number
num = int(input("Enter number: "))

# Initialize reverse number
rev = 0

# Loop until number becomes 0
while num > 0:
    # Extract last digit
    digit = num % 10
    # Add digit to reversed number
    rev = rev * 10 + digit
    # Remove last digit
    num = num // 10

# Display reversed number
print("Reversed number:", rev)
```

```python
# Insert number
num = int(input("Enter number: "))

# Assume number is prime
prime = True

# Check divisibility
for i in range(2, num):
    # If divisible
    if num % i == 0:
        # Not prime
        prime = False
        break

if prime:
    print("Prime number")
else:
    print("Not prime number")
```

```python
# Create a list
numbers = [4, 7, 1, 9, 3]

# Assume first number is largest
largest = numbers[0]

# Check every number
for num in numbers:
    # Compare numbers
    if num > largest:
        # Update largest number
        largest = num

print("Largest number:", largest)
```
